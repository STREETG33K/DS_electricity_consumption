```python
import pandas as p
import matplotlib.pyplot as plt

df = p.read_csv("data/electricityConsumptionAndProductioction.csv")
df = df.set_index("DateTime")
df.index = p.to_datetime(df.index)
df.head(10)
```

```python
production_by_source = df[['Nuclear', 'Wind', 'Hydroelectric', 'Oil and Gas', 'Coal', 'Solar', 'Biomass']].sum()
source_labels = production_by_source.index

plt.figure(figsize=(8, 10))
plt.pie(production_by_source, labels=source_labels, autopct='%1.1f%%', startangle=90)
plt.title('Energy Source Contributions to Total Production')
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you've already read the dataset and processed it as in your previous code
# If not, you can use your previous code to read and prepare the DataFrame 'df'

# Extract the year from the index and add it as a new column
df['Year'] = df.index.year

# Get the unique years in the dataset
unique_years = df['Year'].unique()

# Loop through each year and create a pie chart
for year in unique_years:
    year_df = df[df['Year'] == year]
    production_by_source = year_df[['Nuclear', 'Wind', 'Hydroelectric', 'Oil and Gas', 'Coal', 'Solar', 'Biomass']].sum()
    source_labels = production_by_source.index

    fig = plt.figure(figsize=(8, 10))
    #ax = fig.add_subplot(3, 2, i + 1)
    plt.pie(production_by_source, labels=source_labels, autopct='%1.1f%%', startangle=90)
    plt.title(f'Energy Source Contributions to Total Production - {year}')
    plt.axis('equal') 
    plt.show()
```

```python
scalar_nuclear = 0.5
scalar_wind = 2.5
scalar_solar = 4
scalar_coal = 0.5

df['Nuclear'] *= scalar_nuclear
df['Wind'] *= scalar_wind
df['Solar'] *= scalar_solar
df['Coal'] *= scalar_coal

production_by_source = df[['Nuclear', 'Wind', 'Hydroelectric', 'Oil and Gas', 'Coal', 'Solar', 'Biomass']].sum()
source_labels = production_by_source.index

plt.figure(figsize=(8, 10))
plt.pie(production_by_source, labels=source_labels, autopct='%1.1f%%', startangle=90)
plt.title('Energy Source Contributions to Total Production')
plt.axis('equal')
plt.show()





```
